# US Monthly Sales of Petroleum

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Introduction

This project looks at the sales of Chemicals, Coal, Petrol and Vehicles in the US between 1971 and 1991. The data from 252 rows with each divided up into 12 months.

Website https://datamarket.com/data/set/22nd/us-monthly-sales-of-petroleum-and-related-products-jan-1971-dec-1991#!ds=22nd!2ekn&display=line

```python
df = pd.read_csv('/home/james/Downloads/us-monthly-sales-of-petroleum-an.csv', parse_dates=True)
df.head()
```

```python
len(df['Month'])
```

```python
df['year'] = pd.DatetimeIndex(df['Month']).year
```

Each plot shows the general trend for Chemicals, Coal, Petrol and Vehicles sales. Each X label represents 2.5 years from 1971 to 1991. The vertical lines represent the stability of each point, the larger the vertical line the more unstable the sales were at that time.

```python
plt.figure(figsize=(20,10))
plt.style.available
plt.plot(df.year, df.Chemicals, color='blue')
plt.xticks(rotation=90)
plt.title('Chemical Sales Between 1971 and 1991', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.xlim(1970, 1992)
plt.style.use('ggplot')
plt.show()
```

Chemical sales in the US show a gradual increase over this time period. The vertical line shows the sales over this time remained relitvilty stable, with sales raising from under 5 to just under 25 in the 20 year period.

```python
plt.figure(figsize=(20,10))
plt.plot(df.year, df.Coal, color='red')
plt.xticks(rotation=90)
plt.title('Coal Sales Between 1971 and 1991', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.xlim(1970, 1992)
plt.style.use('ggplot')
plt.show()
```

Coal sales shows a gradual increase in sales between 1971 to 1991. However, unlike Chemical sales the vertical lines show that the sales for coal were very unstable between 1976 and 1984 with large drops in sales around 1977.

```python
plt.figure(figsize=(20,10))
plt.plot(df.year, df.Petrol, color='black')
plt.xticks(rotation=90)
plt.title('Petrol Sales Between 1971 and 1991', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.xlim(1970, 1992)
plt.style.use('ggplot')
plt.show()
```

Petrol sales remained stable up to 1978, there was a significant sales increase from 1979 to 1981, Although, the sales started to decrease with the Iraq - Iran war starting in September in 1980. Also, the sales increase in 1990 may have been a result of Iraq invasion of Kuwait in 1990.

```python
plt.figure(figsize=(20,10))
plt.plot(df.year, df.Vehicles, color='green')
plt.xticks(rotation=90)
plt.title('Vehicles Sales Between 1971 and 1991', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.xlim(1970, 1992)
plt.style.use('ggplot')
plt.show()
```

Vehicles sales started in 1971 started reasonably stable. In 1979-1980 sales decrease possible as a result of higher Petrol sales at around this time. In the 1980s the sales gradual rise although the larger vertical lines indicate that vehicle sales are largely unstable.
